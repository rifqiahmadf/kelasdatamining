```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame({
    'x': [23, 51, 31, 4, 6, 77, 42, 46, 54, 45, 98, 1, 51, 32, 49, 30, 24, 12, 18],
    'y': [7, 12, 29, 35, 41, 57, 68, 75, 81, 90, 4, 19, 21, 39, 47, 56, 60, 79, 82]
})
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dendrogram = sch.dendrogram(sch.linkage(df,method='ward'))
```

```python
plt.scatter(df['x'],df['y'])
```

```python
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
```

```python
y_hc=hc.fit_predict(df)
```

```python
dendrogram = sch.dendrogram(sch.linkage(np.reshape(y_hc, (len(y_hc), 1)),method='ward'))
```

```python
print(y_hc)
```

```python
plt.scatter(df['x'],df['y'], c=y_hc, cmap='rainbow')
```
